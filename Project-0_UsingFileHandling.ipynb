*Another file is also given that contains some queries you'll have answer all those queries*

```python
moviesFile=open(r'C:\Users\sulem\Project_1\movie.txt','r')
```

```python
from collections import defaultdict                #import default dictionary to ignore the key check
dictMovies=defaultdict(list)                       #Intializing default dict
for i in moviesFile:
    tempList=i[:-1].split(',')
    dictMovies['serialNumber'].append(tempList[0])
    dictMovies['movieName'].append(tempList[1])
    dictMovies['releaseYear'].append(tempList[2])
    dictMovies['Rating'].append(tempList[3])
    dictMovies['durationInSec'].append(tempList[4])
moviesFile.close()
```

*In the above loop I have just stored all the data into dicionary(Hash Map) and keys are serialNumber, movieName, releaseYear, Rating, durationInSec*

*in every key there is a list containing all one record of movies as the name*

*ith index in every list shows one record i.e. serialNumber, movieName, releaseYear, Rating, durationInSec and we can print that record as below*

```python
print('Serial Number:       ',dictMovies['serialNumber'][0])
print('Movie Name:          ', dictMovies['movieName'][0])
print('Release Year:        ',dictMovies['releaseYear'][0])
print('Rating of Movie:     ',dictMovies['Rating'][0])
print('Duration in Seconds: ',dictMovies['durationInSec'][0])
```

### Query-1 
**Find the number of movies released between 1950 and 1960.**

```python
countMovies=0
for year in dictMovies['releaseYear']: # Looping through the list stored at key releaseYear
    year=int(year)                     #Type Casting
    if(year>1950 and year<1960):       #Checking whether the current year is greater than 1950 and less than 1960
        countMovies+=1
print('The number of Movies released between 1950 and 1960 is: {}'.format(countMovies)) #printing the countMovies
```

### Query-2
**Find the number of movies having rating more than 4.**

```python
countMovies=0
for rating in dictMovies['Rating']:
    try:
        rating=float(rating)
        if(rating>4):
            countMovies+=1
    except:
        pass
print('The number of movies having rating more than 4 is: {}'.format(countMovies))
```

### Query-4
**Find the number of movies with duration more than 2 hours (7200 seconds)**

```python
countMovies=0
for duration in dictMovies['durationInSec']:
    try:
        duration=int(duration)
        if(duration>7200):
            countMovies+=1
    except:
        pass
print('Number of movies having duration more than 2 hours(7200 seconds) is: {}'.format(countMovies))
```

### Query-5
**Find the list of years and number of movies released each year**

```python
dictReleaseYear=defaultdict(int)
for year in dictMovies['releaseYear']:
    dictReleaseYear[year]+=1
print('List of the year and number of movies released each year: ')
print()
for year, movieCounts in dictReleaseYear.items():
    print(year, movieCounts)
```
